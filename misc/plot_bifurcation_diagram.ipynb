```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
df = pd.read_csv('extrema_beta09.csv')
params = df['Parameter'].values
state = df['Value'].values

plt.scatter(params,state,s=0.1,c="k",label="beta=0.9")

#df = pd.read_csv('extrema_beta16.csv')
df = pd.read_csv('extrema_beta16.csv')
params = df['Parameter'].values
state = df['Value'].values

plt.scatter(params,state,s=0.1,c="r",label="beta=1.6",alpha=0.95)

plt.xlim(-0.2,3.2)
plt.xlabel('Parameter b')
plt.ylabel('State x*')
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(0.1))
plt.legend(frameon=False)
# plt.grid(which='both')
```
