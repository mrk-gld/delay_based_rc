```python
# plot prediction vs target from files
df = pd.read_csv("y_pred.csv", header=None)
data_pred = np.array(df.iloc[:])
plt.plot(data_pred, label='prediction')

df = pd.read_csv("y_test.csv", header=None)
data_test = np.array(df.iloc[:])
plt.plot(data_test, label='target')
plt.legend()
plt.show()

nrmse = np.sqrt(np.mean((data_pred-data_test)**2))/np.std(data_test)
print("NRMSE: ", nrmse)
```
